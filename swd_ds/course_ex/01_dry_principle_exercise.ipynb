**Hint 1 — Group related data**

Instead of separate variables (`student1_name`, `student1_grade_math`, ...), store each student's full data in one dictionary:


**Hint 2 — Extract repeated logic into functions**

Any block of code that looks similar more than once is a candidate for a function. The grade-letter conversion and the report printing are both good targets.

```python

    # your if/elif/else logic here

```

---

**Hint 3 — Replace copy-paste with a loop**

Once your data is in a list and your logic is in functions, a single `for` loop handles every student:

```python
for student in students:

```

---

**Hint 4 — Bonus: finding the top student**

`max()` accepts a `key` function. A `lambda` is a short anonymous function:

```python
top = max(students, key=lambda s: sum(s["grades"].values()) / len(s["grades"]))
```

## Part 5 — Model Solution

> **Only read this after you've made a genuine attempt!**

Run the cell below to see (and execute) the model solution.

````python
# --- MODEL SOLUTION ---

# 1. Grade logic defined ONCE
def get_letter_grade(average):
    if average >= 90:   return "A"
    elif average >= 80: return "B"
    elif average >= 70: return "C"
    else:               return "F"

# 2. Report printing defined ONCE
def print_report(student):
    avg = sum(student["grades"].values()) / len(student["grades"])
    letter = get_letter_grade(avg)
    print(f"--- Report for {student['name']} ---")
    for subject, score in student["grades"].items():
        print(f"  {subject:<10} {score}")
    print(f"  Average: {avg:.1f}  |  Grade: {letter}\n")

# 3. All data in one structured list
students = [
    {"name": "Alice",   "grades": {"Math": 88, "Science": 92, "English": 75}},
    {"name": "Bob",     "grades": {"Math": 70, "Science": 60, "English": 85}},
    {"name": "Charlie", "grades": {"Math": 95, "Science": 89, "English": 91}},
]

# 4. One loop replaces all repetition
for student in students:
    print_report(student)

# BONUS: Top student in one line
top = max(students, key=lambda s: sum(s["grades"].values()) / len(s["grades"]))
print(f"Top student: {top['name']}")
````
