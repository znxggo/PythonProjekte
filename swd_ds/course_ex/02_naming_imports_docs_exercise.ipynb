## Naming · Imports · Documentation — PEP 8 & Beyond

## Why Conventions Matter

Code is read far more often than it is written. Following Python's style conventions — primarily **PEP 8** — makes your code immediately understandable to any Python developer, and signals that you write code meant to be maintained, not just executed.

This notebook focuses on three pillars:

| Pillar | Key Questions |
|---|---|
| **Naming** | Does the name reveal intent? Does it follow the right case style? |
| **Imports** | Are imports ordered, specific, and aliased correctly? |
| **Documentation** | Does every public function/class have a docstring? Are comments explaining *why*, not *what*? |

### Learning Goals
- Apply correct **PEP 8 naming conventions** for variables, functions, classes, and constants
- Write **well-structured imports** (order, specificity, aliases)
- Write **meaningful docstrings** in Google / NumPy style
- Distinguish **good comments** (explain intent) from bad ones (restate code)
- Recognise and fix a full module that violates all three pillars at once

## Quick Reference — PEP 8 Naming Rules

| What | Convention | Example |
|---|---|---|
| Variable | `snake_case` | `pixel_count` |
| Function | `snake_case` | `calculate_average()` |
| Method | `snake_case` | `self.load_data()` |
| Class | `PascalCase` | `ImageProcessor` |
| Constant (module-level) | `UPPER_SNAKE_CASE` | `MAX_RETRIES = 3` |
| Private variable/method | `_leading_underscore` | `_cache = {}` |
| "Dunder" (special) | `__double_underscore__` | `__init__`, `__str__` |
| Module / package | `lowercase` (short) | `utils`, `data_loader` |

**Key name quality rules:**
- Names must reveal **intent** — avoid `x`, `tmp`, `data2`, `foo`
- Avoid abbreviations unless they are universally understood (`url`, `id`, `html`)
- Boolean names should read as questions: `is_valid`, `has_errors`, `can_retry`
- Collection names should be plural: `students`, `error_codes`

## Quick Reference — Import Rules (PEP 8)

**Order:** Imports must appear at the top of the file in three groups, separated by blank lines:
```
1. Standard library   (os, sys, math, datetime ...)
2. Third-party        (numpy, pandas, requests ...)
3. Local / project    (your own modules)
```

**Specificity:** Import only what you need.
```python
# Bad  — pollutes namespace, hides where names come from
from os.path import *

# Good — explicit is better than implicit
from os.path import join, exists
```

**Aliases:** Use standard, universally recognised aliases only.
```python
import numpy as np          # ✅ universal
import pandas as pd         # ✅ universal
import matplotlib.pyplot as plt  # ✅ universal
import numpy as n           # ❌ non-standard, confusing
import pandas as pandas_lib # ❌ verbose
```

**One import per line:**
```python
import os, sys   # ❌ PEP 8 violation
import os        # ✅
import sys       # ✅
```

## Quick Reference — Docstrings & Comments

### Docstrings (PEP 257)
Every **public** module, class, and function should have a docstring.

```python
def calculate_average(values):
    """Return the arithmetic mean of a list of numbers.

    Args:
        values (list[float]): A non-empty list of numeric values.

    Returns:
        float: The arithmetic mean.

    Raises:
        ValueError: If `values` is empty.

    Example:
        >>> calculate_average([10, 20, 30])
        20.0
    """
```

**Rules:**
- First line: one-sentence summary ending with a period.
- Blank line between summary and extended description.
- Document `Args`, `Returns`, and `Raises` for non-trivial functions.
- Use `"""triple quotes"""` even for one-liners.

### Inline Comments
```python
# ❌ Bad — restates the code (adds zero information)
x = x + 1  # increment x by 1

# ✅ Good — explains WHY (adds context a reader couldn't infer)
x = x + 1  # account for the header row in the CSV
```

**Rules:**
- Comments explain **why**, not what.
- Start with a capital letter; end with a period if it is a full sentence.
- Separate inline comments from code with two spaces: `code  # comment`
- Delete commented-out code — use version control instead.

````python
# === EXERCISE 4 SOLUTION — Docstrings ===

def normalise_pixel(pixel_value, max_value=255):
    """Scale a pixel intensity to the [0.0, 1.0] range.

    Args:
        pixel_value (int | float): Raw pixel intensity.
        max_value (int): Maximum intensity value. Defaults to 255.

    Returns:
        float: Normalised value in [0.0, 1.0].

    Raises:
        TypeError: If `pixel_value` is not numeric.
        ValueError: If `pixel_value` is outside [0, max_value].

    Example:
        >>> normalise_pixel(128)
        0.5019607843137255
    """
    if not isinstance(pixel_value, (int, float)):
        raise TypeError("pixel_value must be numeric")
    if not (0 <= pixel_value <= max_value):
        raise ValueError(f"pixel_value must be in [0, {max_value}]")
    return float(pixel_value) / max_value


def get_letter_grade(average):
    """Map a numeric average to a letter grade.

    Uses the standard US grading scale: A (90–100), B (80–89),
    C (70–79), D (60–69), F (below 60).

    Args:
        average (float): Numeric score in the range [0, 100].

    Returns:
        str: A single letter grade ('A', 'B', 'C', 'D', or 'F').

    Example:
        >>> get_letter_grade(85)
        'B'
    """
    if   average >= 90: return "A"
    elif average >= 80: return "B"
    elif average >= 70: return "C"
    elif average >= 60: return "D"
    else:               return "F"


class StudentReport:
    """Represents a single student's grades and generates a summary report."""

    def __init__(self, name, grades):
        """Initialise the report with a student name and grade mapping.

        Args:
            name (str): Full name of the student.
            grades (dict[str, float]): Mapping of subject name to numeric score.
        """
        self.name = name
        self.grades = grades

    def average(self):
        """Return the mean score across all subjects.

        Returns:
            float: Arithmetic mean of all grade values.
        """
        return sum(self.grades.values()) / len(self.grades)

    def summary(self):
        """Return a one-line human-readable summary of the student's performance.

        Returns:
            str: e.g. 'Alice: 85.0 (B)'
        """
        avg = self.average()
        return f"{self.name}: {avg:.1f} ({get_letter_grade(avg)})"


print(normalise_pixel(128))
print(get_letter_grade(85))
r = StudentReport("Alice", {"Math": 88, "Science": 92, "English": 75})
print(r.summary())
````

````python
# === EXERCISE 5 SOLUTION — Full Module Refactor ===

# Violations found (16 total):
#  1.  `import numpy as numpylib` — non-standard alias
#  2.  `import numpy as numpylib, os` — two imports on one line
#  3.  `from math import *` — wildcard import
#  4.  Imports are not grouped (stdlib / third-party / local mixed)
#  5.  `maxS` — should be MAX_SCORE (constant → UPPER_SNAKE_CASE)
#  6.  `minS` — should be MIN_SCORE
#  7.  `Threshold` — mixed case for a constant; should be PASSING_THRESHOLD
#  8.  `imgProc` — class should be PascalCase: `ImageProcessor`
#  9.  `__init__(self, p)` — `p` reveals nothing; should be `pixel_values`
# 10.  `self.done` — unclear; should be `self.is_normalised`
# 11.  `Normalise` — method should be snake_case: `normalise`
# 12.  `calcMn` — should be `calculate_mean`
# 13.  `chk(val)` — function name reveals nothing; should be `is_above_threshold`
# 14.  `proc(lst, fl)` — both function and parameter names meaningless
# 15.  No docstrings on any class, method, or function
# 16.  Inline comments restate code ("store pixels", "loop through pixels", "add each pixel")

# --- Standard library ---
import os
import sys

# --- Third-party ---
import numpy as np
import pandas as pd

MAX_SCORE = 100
MIN_SCORE = 0
PASSING_THRESHOLD = 50


class ImageProcessor:
    """Stores and processes a sequence of raw pixel intensity values."""

    def __init__(self, pixel_values):
        """Initialise the processor with a list of raw pixel values.

        Args:
            pixel_values (list[int | float]): Raw pixel intensities in [0, 255].
        """
        self.pixel_values = pixel_values
        self.is_normalised = False

    def normalise(self):
        """Scale all pixel values to the [0.0, 1.0] range in place."""
        self.pixel_values = [float(p) / 255 for p in self.pixel_values]
        self.is_normalised = True

    def calculate_mean(self):
        """Return the arithmetic mean of the stored pixel values.

        Returns:
            float: Mean pixel value.
        """
        return sum(self.pixel_values) / len(self.pixel_values)


def is_above_threshold(value):
    """Return True if `value` meets or exceeds PASSING_THRESHOLD.

    Args:
        value (float): The value to test.

    Returns:
        bool: True if value >= PASSING_THRESHOLD, False otherwise.
    """
    return value >= PASSING_THRESHOLD


def process_values(values, normalise=False):
    """Optionally normalise a list of values against MAX_SCORE.

    Args:
        values (list[int | float]): Raw numeric values to process.
        normalise (bool): If True, divide each value by MAX_SCORE.
            Defaults to False.

    Returns:
        list[float]: Processed values.
    """
    if normalise:
        return [float(v) / MAX_SCORE for v in values]
    return list(values)


# Smoke test
processor = ImageProcessor([0, 128, 255])
processor.normalise()
print(processor.pixel_values)
print(processor.calculate_mean())
print(is_above_threshold(60))
print(process_values([40, 80, 100], normalise=True))
````
