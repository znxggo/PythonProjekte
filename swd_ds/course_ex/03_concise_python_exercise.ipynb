## Avoiding Verbose Code — Idiomatic Python

### What is "Verbose" Code?

Verbose code does the job, but uses more steps, variables, or lines than necessary.
Python is designed to let you express ideas concisely. Compare:

```python
# Verbose
i = float(i)
image_vector.append(i / 255.0)

# Concise — one expression, no throwaway variable
image_vector.append(float(i) / 255)
```

Concise code is **not** about being clever or hard to read — it's about removing noise so the intent is clearer.

### Learning Goals
- Replace throwaway variables with inline expressions
- Use **list comprehensions** instead of loop-and-append
- Use **ternary expressions** instead of simple if/else blocks
- Use **chained comparisons** and **built-in functions**
- Use **tuple unpacking** and `enumerate` / `zip`
- Use **f-strings** and string methods effectively

## Hints — Reference Card

| Technique | Verbose pattern | Concise pattern |
|---|---|---|
| Inline expression | `y = f(x)` then use `y` once | use `f(x)` directly |
| List comprehension | `for x in d: lst.append(...)` | `[... for x in d]` |
| Filtered comprehension | `if cond: lst.append(x)` | `[x for x in d if cond]` |
| Ternary | `if c: v=a` / `else: v=b` | `v = a if c else b` |
| Default value | `if x: v=x` / `else: v=default` | `v = x or default` |
| Sum | manual `total +=` loop | `sum(data)` |
| Any / All | manual flag + loop | `any(...)` / `all(...)` |
| Indexed loop | `for i in range(len(lst))` | `for i, v in enumerate(lst)` |
| Parallel loop | `lst1[i]`, `lst2[i]` in loop | `for a, b in zip(lst1, lst2)` |
| Unpacking | `x=t[0]; y=t[1]` | `x, y = t` |
| Chained comparison | `x > 0 and x < 10` | `0 < x < 10` |
| Boolean check | `if flag == True` | `if flag` |
| Membership | `x==a or x==b or x==c` | `x in (a, b, c)` |
| Method chaining | `a=s.strip()` then `b=a.upper()` | `s.strip().upper()` |
