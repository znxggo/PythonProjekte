## Replacing Unexplained Literals with Named Constants

### What is a Magic Number?

A **magic number** is a numeric (or string) literal that appears in code without explanation.  
The reader has no idea what it means or where it comes from.

```python
# Magic number — what does 86400 mean?
remaining = total_seconds / 86400

# Named constant — intent is immediately clear
SECONDS_PER_DAY = 86400
remaining = total_seconds / SECONDS_PER_DAY
```

Magic numbers hurt readability in three ways:
- A reader must **guess or research** what the value means
- When the value must change, you have to **hunt down every occurrence**
- The same literal used in two places may mean **two different things**

### Learning Goals
- Recognise magic numbers and magic strings in real code
- Replace them with well-named `UPPER_SNAKE_CASE` constants
- Group related constants using a class or `Enum`
- Understand when a literal is acceptable (e.g. `0`, `1`, `""`, `True`)

## Quick Reference

### When to use a named constant
| Situation | Example literal → constant |
|---|---|
| Domain value with a specific meaning | `255` → `MAX_PIXEL_VALUE` |
| Configuration threshold | `3` → `MAX_LOGIN_ATTEMPTS` |
| Unit conversion factor | `1000` → `METRES_PER_KILOMETRE` |
| Status / category string | `"admin"` → `ROLE_ADMIN` |
| Repeated literal used in multiple places | any → give it one name |

### When a bare literal is fine
- `0` and `1` as initializers or loop boundaries (almost always self-evident)
- `True` / `False` / `None`
- `""` empty string checks
- Simple arithmetic identity values (`/ 2`, `* -1`)

### Grouping related constants
```python
# Option A — module-level constants (simple cases)
MIN_AGE = 18
MAX_AGE = 120

# Option B — class namespace (related group)
class AgeLimit:
    MIN = 18
    MAX = 120

# Option C — Enum (mutually exclusive categories)
from enum import Enum

class UserRole(Enum):
    ADMIN  = "admin"
    EDITOR = "editor"
    VIEWER = "viewer"
```

````python
print("=== Exercise 1a ===")
KMH_TO_MS_FACTOR     = 3.6
MOON_GRAVITY_MS2     = 1.62

def speed_kmh_to_ms(kmh: float) -> float:
    return kmh / KMH_TO_MS_FACTOR

def kinetic_energy(mass_kg: float, speed_ms: float) -> float:
    return 0.5 * mass_kg * speed_ms ** 2  # 0.5 is part of the formula — not magic

def weight_on_moon(mass_kg: float) -> float:
    return mass_kg * MOON_GRAVITY_MS2

print(speed_kmh_to_ms(90))    # 25.0
print(kinetic_energy(70, 10)) # 3500.0
print(weight_on_moon(80))     # 129.6
````

````python
print("=== Exercise 1b ===")
HTTP_SUCCESS_MIN    = 200
HTTP_SUCCESS_MAX    = 300
HTTP_SERVER_ERROR   = 500
HTTP_RATE_LIMITED   = 429
MAX_RETRY_ATTEMPTS  = 3

def is_success(status_code: int) -> bool:
    return HTTP_SUCCESS_MIN <= status_code < HTTP_SUCCESS_MAX

def should_retry(status_code: int, attempt: int) -> bool:
    return status_code >= HTTP_SERVER_ERROR and attempt < MAX_RETRY_ATTEMPTS

def is_rate_limited(status_code: int) -> bool:
    return status_code == HTTP_RATE_LIMITED

print(is_success(200))       # True
print(should_retry(503, 2))  # True
print(is_rate_limited(429))  # True
````

````python
print("=== Exercise 2a ===")
from enum import Enum

class OrderStatus(Enum):
    PENDING   = "pending"
    SHIPPED   = "shipped"
    DELIVERED = "delivered"
    CANCELLED = "cancelled"

def describe_order(status: OrderStatus) -> str:
    messages = {
        OrderStatus.PENDING:   "Your order is awaiting confirmation.",
        OrderStatus.SHIPPED:   "Your order is on its way!",
        OrderStatus.DELIVERED: "Your order has been delivered.",
        OrderStatus.CANCELLED: "Your order was cancelled.",
    }
    return messages.get(status, "Unknown status.")

def can_cancel(status: OrderStatus) -> bool:
    return status in (OrderStatus.PENDING, OrderStatus.SHIPPED)

print(describe_order(OrderStatus.SHIPPED))  # Your order is on its way!
print(can_cancel(OrderStatus.PENDING))      # True
````

````python
print("=== Exercise 2b ===")
IMAGE_EXTENSIONS    = frozenset({"jpg", "jpeg", "png", "gif", "webp"})
DOCUMENT_EXTENSIONS = frozenset({"pdf", "doc", "docx", "txt", "odt"})

def _get_extension(filename: str) -> str:
    return filename.rsplit(".", 1)[-1].lower()

def is_image(filename: str) -> bool:
    return _get_extension(filename) in IMAGE_EXTENSIONS

def is_document(filename: str) -> bool:
    return _get_extension(filename) in DOCUMENT_EXTENSIONS

print(is_image("photo.png"))      # True
print(is_document("report.pdf"))  # True
````

````python
print("=== Exercise 3 ===")
TAX_RATE           = 0.15
MIN_BUYER_AGE      = 18
MIN_CREDIT_SCORE   = 150

def calculate_tax(price: float) -> float:
    return price * TAX_RATE

def final_price(price: float) -> float:
    return price + calculate_tax(price)

def tax_report(prices: list[float]) -> list[str]:
    return [f"${p:.2f} + ${calculate_tax(p):.2f} tax" for p in prices]

def is_eligible_buyer(age: int, credit_score: int) -> bool:
    return age >= MIN_BUYER_AGE and credit_score >= MIN_CREDIT_SCORE

def is_young_buyer(age: int) -> bool:
    return age < MIN_BUYER_AGE

def get_buyer_tier(credit_score: int) -> str:
    return "standard" if credit_score >= MIN_CREDIT_SCORE else "restricted"

prices = [10.0, 25.0, 100.0]
print(tax_report(prices))
print(is_eligible_buyer(20, 200))  # True
print(get_buyer_tier(160))         # standard
````
