# PART A — `*args`

## 1. The Problem `*args` Solves

Suppose you want to write a function that adds numbers together. How many numbers? You don't know in advance.

Without `*args`, you'd have to write a different function for every possible count — which is obviously unworkable:

````python
# ❌ The bad approach — a different function for every count
def add_two(a, b):         return a + b
def add_three(a, b, c):    return a + b + c
def add_four(a, b, c, d):  return a + b + c + d
# ... this never ends

# ✅ The *args approach — one function, any number of arguments
def add(*args):
    return sum(args)

print(add(1, 2))           # 3
print(add(1, 2, 3))        # 6
print(add(1, 2, 3, 4, 5))  # 15
print(add())               # 0  — zero arguments also works!
````

## 2. How `*args` Works — The Mechanics

When Python sees `*args` in a function definition:
1. It collects **all extra positional arguments** into a **tuple**.
2. The name `args` is just a convention — you could write `*numbers` or `*items`. The `*` is what matters.
3. Because it's a tuple, you can loop over it, index it, pass it to `len()`, etc.
4. If no extra arguments are passed, `args` is simply an **empty tuple** `()`.

````python
# Inspecting what *args actually contains

def inspect(*args):
    print(f"Type : {type(args)}")
    print(f"Value: {args}")
    print(f"Length: {len(args)}")
    for i, arg in enumerate(args):
        print(f"  args[{i}] = {arg!r}  (type: {type(arg).__name__})")

print("--- Three arguments ---")
inspect(10, "hello", [1, 2, 3])

print("\n--- No arguments ---")
inspect()
````

````python
# The name after * is just a label — args, numbers, items, anything works

def multiply(*numbers):
    result = 1
    for n in numbers:
        result *= n
    return result

def join_words(*words):
    return " ".join(words)

print(multiply(2, 3, 4))              # 24
print(join_words("Hello", "world"))   # Hello world
print(join_words("one", "two", "three", "four"))  # one two three four
````

````python
# Since args is a tuple, you can use all tuple operations on it

def stats(*numbers):
    if not numbers:
        return None
    return {
        "count" : len(numbers),
        "sum"   : sum(numbers),
        "min"   : min(numbers),
        "max"   : max(numbers),
        "mean"  : sum(numbers) / len(numbers),
        "first" : numbers[0],          # index like a tuple
        "last"  : numbers[-1],
    }

result = stats(4, 7, 2, 9, 1, 5)
for key, value in result.items():
    print(f"  {key:6}: {value}")
````

## 3. Combining `*args` with Regular Parameters

You can mix `*args` with regular (positional) parameters — but the order matters:

```python
def func(regular_params, *args):
```

- Regular parameters come **first** and are filled left-to-right as usual.
- `*args` collects **everything that's left over**.

⚠️ You **cannot** put a regular parameter after `*args` without making it keyword-only (covered in Part C).

````python
# Regular parameters are filled first; *args gets the rest

def greet(greeting, *names):
    for name in names:
        print(f"{greeting}, {name}!")

greet("Hello", "Alice")                    # greeting="Hello", names=("Alice",)
print()
greet("Good morning", "Alice", "Bob", "Charlie")  # names=("Alice","Bob","Charlie")
````

````python
# Real-world pattern: a required "label" followed by any number of values

def log(level, *messages):
    prefix = f"[{level.upper()}]"
    for msg in messages:
        print(f"{prefix} {msg}")

log("info", "Server started")
log("warning", "Disk at 80%", "CPU at 95%")
log("error", "Connection lost", "Retrying in 5s", "Attempt 1/3")
````

````python
# Default values + *args

def repeat(times=1, *items):
    for item in items:
        print((str(item) + " ") * times)

repeat(3, "echo", "test")   # times=3, items=("echo", "test")
print()

# ⚠️ Watch out: the first positional arg ALWAYS goes to `times`, not items
repeat(2, "hello")   # times=2, items=("hello",)
````

## 4. Unpacking with `*` in Function Calls

The `*` operator does double duty:
- In a **function definition**: collects extra positional args into a tuple (`*args`)
- In a **function call**: **unpacks** an iterable into individual positional arguments

These are two sides of the same coin.

````python
# Unpacking a list/tuple into positional arguments

def add(a, b, c):
    return a + b + c

numbers = [1, 2, 3]

# Without unpacking — TypeError!
# add(numbers)   ← this passes ONE argument (a list), not three

# With unpacking — perfect
print(add(*numbers))   # same as add(1, 2, 3)

# Works with any iterable
print(add(*range(3)))          # add(0, 1, 2) → 3
print(add(*(10, 20, 30)))      # tuple unpacking → 60
````

````python
# Combining unpacking with regular arguments

def describe(title, *items):
    print(f"{title}:")
    for item in items:
        print(f"  - {item}")

fruits = ["apple", "banana", "cherry"]

describe("Fruit basket", *fruits)
# equivalent to: describe("Fruit basket", "apple", "banana", "cherry")
````

````python
# * unpacking is also useful for merging iterables

a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

merged = [*a, *b, *c]   # unpack all three into a new list
print(merged)           # [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Also works for sets and tuples
combined_set = {*a, *b}   # {1, 2, 3, 4, 5, 6}
print(combined_set)

# And for assignment unpacking
first, *rest = [10, 20, 30, 40, 50]
print(f"first={first}, rest={rest}")

*start, last = [10, 20, 30, 40, 50]
print(f"start={start}, last={last}")
````

# PART B — `**kwargs`

## 5. The Problem `**kwargs` Solves

`**kwargs` solves the same problem as `*args` — but for **keyword arguments** (the `name=value` kind).

Imagine building a function that creates an HTML element. You don't know in advance which attributes the caller might want to set:

````python
# ❌ Trying to pre-define every possible HTML attribute — impossible
def make_tag_bad(tag, id=None, class_=None, style=None, href=None, src=None):
    # Can't cover all attributes: data-*, aria-*, custom attributes, etc.
    pass

# ✅ **kwargs collects any keyword arguments the caller passes
def make_tag(tag, content="", **attrs):
    attr_str = " ".join(f'{k}="{v}"' for k, v in attrs.items())
    if attr_str:
        return f"<{tag} {attr_str}>{content}</{tag}>"
    return f"<{tag}>{content}</{tag}>"

print(make_tag("p", "Hello world"))
print(make_tag("a", "Click me", href="https://python.org", target="_blank"))
print(make_tag("img", src="photo.jpg", alt="A photo", width="300"))
print(make_tag("div", "Content", id="main", style="color:red", data_value="42"))
````

## 6. How `**kwargs` Works — The Mechanics

When Python sees `**kwargs` in a function definition:
1. It collects **all extra keyword arguments** into a **dict** (key = argument name, value = argument value).
2. The name `kwargs` is just a convention — you could write `**options` or `**config`. The `**` is what matters.
3. Because it's a dict, you can use `.items()`, `.get()`, `.keys()`, etc.
4. If no extra keyword arguments are passed, `kwargs` is simply an **empty dict** `{}`.

````python
# Inspecting what **kwargs actually contains

def inspect(**kwargs):
    print(f"Type : {type(kwargs)}")
    print(f"Value: {kwargs}")
    print(f"Keys : {list(kwargs.keys())}")
    for key, value in kwargs.items():
        print(f"  {key!r:15} → {value!r}  (type: {type(value).__name__})")

print("--- Three keyword arguments ---")
inspect(name="Alice", age=30, active=True)

print("\n--- No keyword arguments ---")
inspect()
````

````python
# Since kwargs is a dict, use all dict operations

def configure(**options):
    # Provide defaults using .get()
    host    = options.get("host",    "localhost")
    port    = options.get("port",    8080)
    debug   = options.get("debug",   False)
    timeout = options.get("timeout", 30)

    print(f"Server config:")
    print(f"  {host}:{port}  debug={debug}  timeout={timeout}s")

    # Warn about any unrecognised options
    known = {"host", "port", "debug", "timeout"}
    unknown = set(options.keys()) - known
    if unknown:
        print(f" Unrecognised options: {unknown}")

configure()
print()
configure(host="example.com", port=443, debug=True)
print()
configure(port=9000, typo_option=True)  # triggers the warning
````

````python
# kwargs preserves insertion order (Python 3.7+)
# and supports all dict methods

def show_profile(**info):
    print("Profile:")
    for key, value in info.items():   # items() iteration
        print(f"  {key.replace('_', ' ').title():15}: {value}")

show_profile(
    first_name="Jane",
    last_name="Doe",
    age=28,
    job_title="Software Engineer",
    years_experience=5,
)
````

## 7. Combining `**kwargs` with Other Parameters

The full parameter order rule in Python:

```
def func(positional, *args, keyword_only, **kwargs):
```

We'll cover the full rule in Part C. For now, the key point:
- `**kwargs` always comes **last**.
- Regular parameters and `*args` are filled first; `**kwargs` collects what's left.

````python
# Mixing regular params with **kwargs

def create_user(username, email, **extra_fields):
    user = {
        "username": username,
        "email":    email,
    }
    user.update(extra_fields)   # merge any extra fields in
    return user

# Minimal call — only required args
u1 = create_user("alice", "alice@example.com")
print(u1)

# Extended call — extra fields go into **kwargs
u2 = create_user(
    "bob", "bob@example.com",
    age=25, role="admin", verified=True
)
print(u2)
````

````python
# Mixing *args AND **kwargs in one function

def event(name, *attendees, **details):
    print(f"Event : {name}")
    print(f"Guests: {', '.join(attendees) if attendees else 'none'}")
    if details:
        print("Details:")
        for k, v in details.items():
            print(f"  {k}: {v}")
    print()

event("Team Lunch")
event("Conference", "Alice", "Bob", "Charlie")
event("Workshop", "Diana", "Eve",
      date="2025-09-01", location="Room 4B", capacity=20)
````

````python
# Forwarding kwargs to another function — a very common pattern

def styled_print(text, **style):
    """A wrapper around print() that supports simple ANSI styling."""
    codes = []
    if style.get("bold"):   codes.append("1")
    if style.get("italic"): codes.append("3")
    color_map = {"red": "31", "green": "32", "blue": "34", "yellow": "33"}
    if "color" in style:
        codes.append(color_map.get(style["color"], "0"))

    # Forward any remaining print() kwargs like end=, sep=
    print_kwargs = {k: v for k, v in style.items()
                    if k in ("end", "sep", "file", "flush")}

    if codes:
        ansi = f"\033[{';'.join(codes)}m"
        print(f"{ansi}{text}\033[0m", **print_kwargs)
    else:
        print(text, **print_kwargs)

styled_print("Normal text")
styled_print("Bold text",        bold=True)
styled_print("Bold green text",  bold=True, color="green")
styled_print("Warning!",         bold=True, color="yellow")
styled_print("Inline: ",         bold=True, color="blue", end="")
styled_print("continues here",   italic=True)
````

## 8. Unpacking with `**` in Function Calls

Just like `*` can unpack a list into positional arguments, `**` can unpack a dict into keyword arguments.

This is useful for:
- Passing config dicts as function arguments
- Merging dicts
- Building arguments dynamically before calling a function

````python
# Unpacking a dict as keyword arguments

def greet(first_name, last_name, title=""):
    if title:
        return f"Hello, {title} {last_name}!"
    return f"Hello, {first_name} {last_name}!"

person = {"first_name": "Jane", "last_name": "Doe", "title": "Dr."}

# Without unpacking — passes a dict as one argument (TypeError!)
# greet(person)  ← wrong

# With unpacking — maps dict keys to parameter names
print(greet(**person))   # Hello, Dr. Doe!

# You can mix ** with regular arguments
base = {"last_name": "Smith"}
print(greet("John", **base))       # Hello, John Smith!
print(greet(**base, first_name="Bob"))  # Hello, Bob Smith!
````

````python
# ** unpacking is extremely useful for merging dicts

defaults = {"color": "blue", "size": 12,  "bold": False}
overrides = {"color": "red",  "italic": True}

# Merge: overrides win over defaults
merged = {**defaults, **overrides}
print(merged)
# {"color": "red", "size": 12, "bold": False, "italic": True}

# Add a single new key while spreading an existing dict
updated = {**merged, "underline": True}
print(updated)
````

````python
# Building arguments programmatically — great for testing and config

def connect(host, port, user, password, timeout=30):
    print(f"Connecting to {user}@{host}:{port} (timeout={timeout}s)")

# You might build this dict from environment variables, a config file, etc.
prod_config = {
    "host":     "db.prod.example.com",
    "port":     5432,
    "user":     "app_user",
    "password": "s3cr3t",
    "timeout":  10,
}

dev_config = {
    "host":     "localhost",
    "port":     5432,
    "user":     "dev",
    "password": "dev",
}

connect(**prod_config)
connect(**dev_config)   # uses default timeout=30
````

# PART C — Putting It All Together

## 9. The Full Parameter Order Rule

Python enforces a strict order for all parameter types:

```
def func(
    positional,                  # 1. Regular positional parameters
    /,                           # 2. (optional) positional-only marker
    positional_or_keyword,       # 3. Regular params (can be passed either way)
    *args,                       # 4. Extra positional → tuple
    keyword_only,                # 5. Keyword-only params (must use name=)
    **kwargs                     # 6. Extra keyword → dict
):
```

In practice, you'll mostly use combinations of 1, 3, 4, and 6.

````python
# Demonstrating the full order

def full_demo(a, b, *args, sep=" ", **kwargs):
    """
    a, b     → required positional
    *args    → extra positional (tuple)
    sep      → keyword-only with default (comes AFTER *args)
    **kwargs → extra keyword (dict)
    """
    print(f"a={a!r}, b={b!r}")
    print(f"args={args}")
    print(f"sep={sep!r}")
    print(f"kwargs={kwargs}")

print("--- Basic call ---")
full_demo(1, 2)

print("\n--- With extras ---")
full_demo(1, 2, 3, 4, 5, sep="-", color="red", width=10)
````

````python
# Keyword-only parameters: anything AFTER *args must be passed by name

def write_file(filename, *lines, encoding="utf-8", overwrite=False):
    """
    filename  → regular positional
    *lines    → all the content lines
    encoding  → keyword-only (can't be passed positionally)
    overwrite → keyword-only
    """
    print(f"Writing '{filename}' ({encoding}, overwrite={overwrite}):")
    for line in lines:
        print(f"  {line}")

write_file("output.txt", "Line 1", "Line 2", "Line 3")
print()
write_file("data.csv", "id,name", "1,Alice", encoding="latin-1", overwrite=True)

# This would be a TypeError — encoding is keyword-only after *args:
# write_file("f.txt", "hello", "utf-16")  ← "utf-16" goes into *lines, not encoding!
````

````python
# Using * alone (no name) to force keyword-only arguments without *args

def send_email(to, subject, *, cc=None, bcc=None, priority="normal"):
    """
    The bare * means: everything after this must be a keyword argument.
    This prevents accidental positional passing of cc/bcc/priority.
    """
    print(f"To:       {to}")
    print(f"Subject:  {subject}")
    if cc:  print(f"CC:       {cc}")
    if bcc: print(f"BCC:      {bcc}")
    print(f"Priority: {priority}")

send_email("alice@example.com", "Hello")
print()
send_email("bob@example.com", "Meeting", cc="boss@example.com", priority="high")

# send_email("x", "y", "cc@x.com")  ← TypeError! cc must be keyword argument
````

## 10.Real-World Patterns

Here are the most common idioms you'll encounter in real codebases.

````python
# Pattern 1: Wrapper / Decorator functions (will be covered in detail in the generators/decorators tutorial)
# *args and **kwargs let you write wrappers that work with ANY function.

import functools, time

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):        # ← accepts ANYTHING the original accepts
        t0 = time.perf_counter()
        result = func(*args, **kwargs)   # ← forwards everything unchanged
        elapsed = time.perf_counter() - t0
        print(f"⏱  {func.__name__}() took {elapsed:.4f}s")
        return result
    return wrapper

@timer
def slow_add(a, b, delay=0.1):
    time.sleep(delay)
    return a + b

@timer
def greet(*names, greeting="Hello"):
    return [f"{greeting}, {n}!" for n in names]

print(slow_add(3, 4))
print(greet("Alice", "Bob", greeting="Hi"))
````

````python
# Pattern 2: Superclass __init__ forwarding
# When extending a class, forward unknown kwargs to the parent.

class Widget:
    def __init__(self, width=100, height=50, color="white"):
        self.width  = width
        self.height = height
        self.color  = color

    def __repr__(self):
        return (f"{type(self).__name__}("
                f"width={self.width}, height={self.height}, color={self.color!r})")

class Button(Widget):
    def __init__(self, label, **kwargs):        # capture widget options
        super().__init__(**kwargs)              # forward them to Widget
        self.label = label

    def __repr__(self):
        return f"Button(label={self.label!r}, {super().__repr__()})"

class IconButton(Button):
    def __init__(self, icon, **kwargs):         # capture all options
        super().__init__(**kwargs)              # forward to Button → Widget
        self.icon = icon

b  = Button("Save", width=120, color="blue")
ib = IconButton("💾", label="Save", width=120, color="blue", height=40)

print(b)
print(ib)
````

````python
# Pattern 3: Building SQL/API queries dynamically

def build_query(table, *columns, **filters):
    """
    build_query("users", "name", "email", age=30, active=True)
    → SELECT name, email FROM users WHERE age = 30 AND active = True
    """
    col_str    = ", ".join(columns) if columns else "*"
    where_parts = [f"{k} = {v!r}" for k, v in filters.items()]
    query = f"SELECT {col_str} FROM {table}"
    if where_parts:
        query += " WHERE " + " AND ".join(where_parts)
    return query

print(build_query("users"))
print(build_query("users", "name", "email"))
print(build_query("users", "name", "email", age=30))
print(build_query("orders", "id", "total", status="shipped", paid=True))
````

````python
# Pattern 4: Combining * and ** unpacking in one call

def draw(shape, *points, **style):
    print(f"Drawing {shape}: points={points}, style={style}")

shape  = "polygon"
pts    = ((0,0), (1,0), (1,1))   # tuple of points
styles = {"color": "red", "fill": True, "stroke": 2}

draw(shape, *pts, **styles)
# equivalent to:
# draw("polygon", (0,0), (1,0), (1,1), color="red", fill=True, stroke=2)
````

## Summary Cheat Sheet

### Quick Reference

| | `*args` | `**kwargs` |
|---|---|---|
| **In definition** | Collects extra *positional* args | Collects extra *keyword* args |
| **Type inside** | `tuple` | `dict` |
| **In a call** | Unpacks an iterable as positional args | Unpacks a dict as keyword args |
| **Position** | After regular params | Always last |
| **Name** | Any name after `*` | Any name after `**` |

### The Full Parameter Order
```python
#  ①regular   ②positional-only  ③regular   ④extra-positional  ⑤keyword-only  ⑥extra-keyword
def func(pos, /,                 normal,    *args,             kw_only=1,     **kwargs):
    pass
# Most common in practice:
def func(required, optional=default, *args, keyword_only=default, **kwargs):
    pass
```

### Patterns Cheat Sheet

```python
# ── *args ────────────────────────────────────────────────────────────

# Collect any number of positional args
def total(*numbers): return sum(numbers)

# Mix with regular params (regular comes first)
def log(level, *messages): ...

# Unpack a list/tuple into positional args
args = [1, 2, 3]
func(*args)         # same as func(1, 2, 3)

# Merge lists
combined = [*list_a, *list_b]

# Unpack assignment
first, *rest = [1, 2, 3, 4]     # first=1, rest=[2,3,4]
*start, last = [1, 2, 3, 4]     # start=[1,2,3], last=4

# ── **kwargs ─────────────────────────────────────────────────────────

# Collect any number of keyword args
def configure(**options): host = options.get("host", "localhost")

# Mix with regular params (**kwargs always last)
def create(name, age, **extra): ...

# Unpack a dict into keyword args
config = {"host": "x", "port": 80}
func(**config)      # same as func(host="x", port=80)

# Merge dicts (last one wins on duplicate keys)
merged = {**defaults, **overrides}

# ── *args + **kwargs together ─────────────────────────────────────────

# Universal wrapper — accepts and forwards anything
def wrapper(*args, **kwargs):
    return original_func(*args, **kwargs)

# Mix of both in a call
positionals = [1, 2, 3]
keywords    = {"color": "red", "size": 10}
func(*positionals, **keywords)
```

### Common Gotchas

```python
# ❌ *args is a TUPLE, not a list — it's immutable
def f(*args): args.append(1)   # AttributeError!
# ✅ Convert first
def f(*args): list(args).append(1)

# ❌ kwargs keys must be valid Python identifiers
func(**{"my-key": 1})   # SyntaxError-like at runtime — hyphens not allowed
# ✅ Use underscores
func(**{"my_key": 1})

# ❌ Passing *args after regular args that were already filled
def f(a, b): ...
f(1, *[2, 3])   # TypeError: too many positional arguments

# ❌ Duplicate keyword arguments
d = {"color": "red"}
f(color="blue", **d)   # TypeError: got multiple values for 'color'
```
