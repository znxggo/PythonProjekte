## 1. Defining a Class

A **class** is a blueprint for creating objects. Think of it as a cookie cutter — the class is the cutter, and objects are the cookies.

```python
class ClassName:
    # attributes and methods go here
    pass
```

An **object** (also called an *instance*) is created by calling the class like a function:
```python
my_object = ClassName()
```

````python
# Example: Defining a simple class and creating instances

class Dog:
    pass  # An empty class — valid Python!

# Creating two Dog instances
dog1 = Dog()
dog2 = Dog()

print(type(dog1))       # <class '__main__.Dog'>
print(dog1 == dog2)     # False — different objects in memory
print(isinstance(dog1, Dog))  # True
````

## 2. The `__init__` Method (Constructor)

`__init__` is called automatically when an object is created. It lets you set up initial values.

- `self` refers to **the current instance** being created.
- Parameters after `self` are values passed when creating the object.

```python
class Person:
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age = age     # instance attribute
```

````python
# Example: Using __init__ to initialize attributes

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Alice", 30)
p2 = Person("Bob", 25)

print(p1.name, p1.age)  # Alice 30
print(p2.name, p2.age)  # Bob 25
````

````python
# Example: Default parameter values in __init__

class Product:
    def __init__(self, name, price, in_stock=True):
        self.name = name
        self.price = price
        self.in_stock = in_stock

p = Product("Laptop", 999.99)
print(p.name, p.price, p.in_stock)  # Laptop 999.99 True

p2 = Product("Old Phone", 49.99, in_stock=False)
print(p2.name, p2.in_stock)  # Old Phone False
````

## 3. Instance Attributes vs Class Attributes

| | **Instance Attribute** | **Class Attribute** |
|---|---|---|
| **Defined in** | `__init__` using `self` | Directly inside the class body |
| **Belongs to** | Each individual object | The class itself (shared) |
| **Access** | `self.attr` or `obj.attr` | `ClassName.attr` or `self.attr` |

Class attributes are shared across **all** instances unless overridden on the instance.

````python
# Example: Class attribute vs Instance attribute

class Dog:
    species = "Canis lupus familiaris"  # Class attribute — shared by all dogs

    def __init__(self, name, breed):
        self.name = breed   # Instance attribute — unique per dog
        self.breed = breed  # Instance attribute

d1 = Dog("Rex", "German Shepherd")
d2 = Dog("Bella", "Labrador")

print(d1.species)       # Canis lupus familiaris (from class)
print(d2.species)       # Same!
print(Dog.species)      # Access directly from class

# Overriding on an instance doesn't affect the class or other instances
d1.species = "Unknown"
print(d1.species)   # Unknown  (instance-level override)
print(d2.species)   # Canis lupus familiaris (unchanged)
````

````python
# Example: Using a class attribute as a counter

class Employee:
    count = 0  # Tracks how many employees were created

    def __init__(self, name):
        self.name = name
        Employee.count += 1  # Increment the class-level counter

e1 = Employee("Alice")
e2 = Employee("Bob")
e3 = Employee("Charlie")

print(f"Total employees: {Employee.count}")  # 3
````

## 4. Instance Methods

**Instance methods** are functions defined inside a class that operate on instance data.
- They always take `self` as their first parameter.
- Called on an object: `obj.method()`

Methods can read and modify the object's attributes.

````python
# Example: Instance methods

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")

    def get_balance(self):
        return self.balance

account = BankAccount("Alice", 100)
account.deposit(50)
account.withdraw(30)
account.withdraw(200)
print("Final balance:", account.get_balance())
````

## 5. Class Methods & Static Methods

### `@classmethod`
- Takes `cls` (the class itself) as the first argument instead of `self`.
- Can access and modify **class-level** data.
- Often used as **alternative constructors**.

### `@staticmethod`
- Takes **no** `self` or `cls` argument.
- A plain function namespaced inside a class.
- Useful for utility/helper functions related to the class.

````python
# Example: @classmethod and @staticmethod

class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    @classmethod
    def from_string(cls, date_string):
        """Alternative constructor: create a Date from 'YYYY-MM-DD'."""
        year, month, day = map(int, date_string.split("-"))
        return cls(year, month, day)  # calls Date(year, month, day)

    @staticmethod
    def is_valid_month(month):
        """Utility: check if a month number is valid."""
        return 1 <= month <= 12

    def __str__(self):
        return f"{self.year}-{self.month:02d}-{self.day:02d}"


# Regular constructor
d1 = Date(2024, 6, 15)
print(d1)  # 2024-06-15

# Alternative constructor via classmethod
d2 = Date.from_string("2025-03-01")
print(d2)  # 2025-03-01

# Static method — no instance needed
print(Date.is_valid_month(13))  # False
print(Date.is_valid_month(7))   # True
````

## 6. Inheritance

**Inheritance** lets a class (child/subclass) acquire attributes and methods from another class (parent/superclass).

```python
class Animal:        # Parent
    ...

class Dog(Animal):   # Child — inherits from Animal
    ...
```

- The child gets **all** parent methods and attributes.
- The child can add its **own** attributes and methods.
- Use `super()` to call the parent's `__init__`.

````python
# Example: Basic inheritance

class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def speak(self):
        print(f"{self.name} says: {self.sound}!")

    def breathe(self):
        print(f"{self.name} breathes air.")


class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Woof")  # Call parent __init__
        self.breed = breed             # Add new attribute

    def fetch(self):
        print(f"{self.name} fetches the ball!")


class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Meow")

    def purr(self):
        print(f"{self.name} purrs...")


rex = Dog("Rex", "Labrador")
whiskers = Cat("Whiskers")

rex.speak()       # Inherited from Animal
rex.breathe()     # Inherited from Animal
rex.fetch()       # Dog-specific
whiskers.speak()  # Inherited from Animal
whiskers.purr()   # Cat-specific

print(isinstance(rex, Dog))     # True
print(isinstance(rex, Animal))  # True — a Dog IS an Animal
````

## 7. Method Overriding & `super()`

**Method overriding** means a child class redefines a method from the parent.
- The child version **replaces** the parent version for that class.
- Use `super().method()` to also run the **parent's version** of a method.

This is how you customize inherited behavior.

````python
# Example: Method overriding

class Notification:
    def __init__(self, message):
        self.message = message

    def send(self):
        print(f"Sending notification: {self.message}")


class EmailNotification(Notification):
    def __init__(self, message, recipient):
        super().__init__(message)  # runs Notification.__init__
        self.recipient = recipient

    def send(self):  # OVERRIDES parent's send()
        print(f"Sending EMAIL to {self.recipient}: {self.message}")


class SMSNotification(Notification):
    def __init__(self, message, phone):
        super().__init__(message)
        self.phone = phone

    def send(self):  # OVERRIDES parent's send()
        super().send()  # Also runs the parent version first
        print(f"Sending SMS to {self.phone}: {self.message}")


n = Notification("Hello")
n.send()  # Generic

e = EmailNotification("Reset your password", "user@example.com")
e.send()  # Email-specific

s = SMSNotification("Your code: 1234", "+1-555-0000")
s.send()  # Runs parent first, then child
````

## 8. Special (Dunder) Methods

**Dunder methods** (double-underscore) let you customize how Python's built-in operations behave for your objects.

| Method | Triggered by |
|---|---|
| `__str__` | `str(obj)` / `print(obj)` |
| `__repr__` | `repr(obj)` / debug display |
| `__len__` | `len(obj)` |
| `__eq__` | `obj1 == obj2` |
| `__lt__` | `obj1 < obj2` |
| `__add__` | `obj1 + obj2` |
| `__contains__` | `item in obj` |

````python
# Example: Dunder methods

class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        """Human-friendly string."""
        return f"'{self.title}' by {self.author}"

    def __repr__(self):
        """Developer/debug string."""
        return f"Book('{self.title}', '{self.author}', {self.pages})"

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.title == other.title and self.author == other.author

    def __lt__(self, other):
        return self.pages < other.pages  # shorter books are "less"


b1 = Book("1984", "Orwell", 328)
b2 = Book("Dune", "Herbert", 688)
b3 = Book("1984", "Orwell", 328)

print(b1)           # __str__
print(repr(b1))     # __repr__
print(len(b1))      # __len__
print(b1 == b3)     # __eq__ → True
print(b1 == b2)     # __eq__ → False
print(b1 < b2)      # __lt__ → True (328 < 688)
````

## Summary Cheat Sheet

```python
class MyClass(ParentClass):                    # Define / inherit
    class_attr = "shared"                      # Class attribute

    def __init__(self, x):                     # Constructor
        super().__init__()                     # Call parent init
        self.x = x                             # Instance attribute

    def instance_method(self):                 # Instance method
        return self.x

    @classmethod
    def class_method(cls):                     # Class method
        return cls.class_attr

    @staticmethod
    def static_method(a, b):                   # Static method
        return a + b

    def __str__(self):                         # Dunder: print()
        return f"MyClass({self.x})"

    def __len__(self):                         # Dunder: len()
        return self.x

    def __eq__(self, other):                   # Dunder: ==
        return self.x == other.x
```

**Key concepts recap:**
- A **class** is a blueprint; an **object** is an instance of it.
- `__init__` runs automatically on creation; `self` is the instance.
- **Class attributes** are shared; **instance attributes** are per-object.
- **Instance methods** use `self`; **class methods** use `cls`; **static methods** use neither.
- **Inheritance** lets child classes reuse parent code. `super()` accesses the parent.
- **Method overriding** lets a child replace a parent's method.
- **Dunder methods** hook into Python operators and built-in functions.
